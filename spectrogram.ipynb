```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram


# Load the WAV file
file_path = '/Users/isab7070/Downloads/wav files/1-us.wav'
output_path = '/Users/isab7070/Downloads/wav files/spectrogram_1-us.png'  # Adjust the file path and format
sample_rate, data = wavfile.read(file_path)

# Create a spectrogram
frequencies, times, Sxx = spectrogram(data, fs=sample_rate)

# Plot the spectrogram
plt.figure(figsize=(12, 8))  # Adjust the figure size as needed
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='auto')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('SPECTOGRAM (ENG US): Hey, have you heard the latest gist about the party next weekend? It\'s gonna be lit!')
plt.colorbar(label='Intensity [dB]')

# Save the plot in HD
plt.savefig(output_path, dpi=300, bbox_inches='tight')  # Adjust dpi as needed
plt.show()
```
